```python
def railfence_encrypt(plaintext, key):
    if key <= 1:
        return plaintext
    
    fence = ['' for _ in range(key)]
    direction = 1  # 1 for down, -1 for up
    row = 0
    
    for char in plaintext:
        fence[row] += char
        row += direction
        
        if row == key - 1 or row == 0:
            direction = -direction
    
    encrypted_text = ''.join(fence)
    return encrypted_text

def railfence_decrypt(ciphertext, key):
    if key <= 1:
        return ciphertext
    
    fence = ['' for _ in range(key)]
    direction = 1
    row = 0
    
    for char in ciphertext:
        fence[row] += 'X'  # Placeholder for rail positions
        row += direction
        
        if row == key - 1 or row == 0:
            direction = -direction
    
    fence[row] += 'X' * (len(ciphertext) - sum(len(rail) for rail in fence))
    
    # Reconstruct the plaintext from the fence
    plaintext = ''
    row = 0
    col = 0
    
    for _ in range(len(ciphertext)):
        plaintext += fence[row][col]
        row += direction
        col += 1
        
        if row == key - 1 or row == 0:
            direction = -direction
    
    return plaintext

# Example usage:
plaintext = "HELLO, WORLD!"
key = 3
print("Plain Text:", plaintext)
print("Key:", key)

# Encryption
encrypted_text = railfence_encrypt(plaintext, key)
print("Encrypted Text:", encrypted_text)

# Decryption
decrypted_text = railfence_decrypt(encrypted_text, key)
print("Decrypted Text:", decrypted_text)
```
