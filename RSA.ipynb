```python
def con_to_num(name):
    res=[]
    arr = list(name)
    for i in range(len(arr)):
        res.append(ord(arr[i]))
    return res

def con_to_char(arr):
    ress=[]
    res=''
    for i in range(len(arr)):
        ress.append(chr(arr[i]))
        res+=ress[i]
    return res

def isprime(n, i=2):
    if n<=2:
        return n==2
    if n%i==0:
        return False
    if i*i>n:
        return True
    return isprime(n,i+1)

def Key_Generation():
    p,q=5,29
    n=p*q
    t=(p-1)*(q-1)
    for i in range(t-1,2,-1):
        if (isprime(i)):
            if (i%t!=0):
                E = i
                break
    for i in range(t,1000):
        if ((i*E)%t==1):
            D=i
            break
    return n,E,D

def encrypt(num, pub, n):
    Cipher = (num**pub)%n
    return Cipher

def decrypt(num, pri, n):
    Message = (num**pri)%n
    return Message


#Numeric Input
Message = 56
product, public_key, private_key = Key_Generation()
Cipher = encrypt(Message,public_key,product)
Decipher = decrypt(Cipher,private_key,product)
print("Message: ", Message)
print("Public Key: ", public_key)
print("Private Key: ", private_key)
print()
print("Cipher Text: ", Cipher)
print("Original Text: ", Decipher)
print()

#Text Input
Message = "A"
product, public_key, private_key = Key_Generation()
Arr = con_to_num(Message)
CipherArr = []
DecipherArr = []
for i in range(len(Arr)):
    CipherArr.append(encrypt(Arr[i],public_key,product))
Cipher = con_to_char(CipherArr)
for i in range(len(CipherArr)):
    DecipherArr.append(decrypt(CipherArr[i],private_key, product))
Decipher=con_to_char(DecipherArr)
print("Message: ", Message)
print("Public Key: ", public_key)
print("Private Key: ", private_key)
print()
print("Cipher Text: ", Cipher, CipherArr)
print("Original Text: ", Decipher, DecipherArr)
print()

#File Input
f = open("D:\Program_Codes\Message.txt",'r')
Message = f.read()
product, public_key, private_key = Key_Generation()
Arr = con_to_num(Message)
CipherArr = []
DecipherArr = []
for i in range(len(Arr)):
    CipherArr.append(encrypt(Arr[i],public_key,product))
Cipher = con_to_char(CipherArr)
for i in range(len(CipherArr)):
    DecipherArr.append(decrypt(CipherArr[i],private_key, product))
Decipher=con_to_char(DecipherArr)
print("Message: ", Message)
print("Public Key: ", public_key)
print("Private Key: ", private_key)
print()
print("Cipher Text: ", Cipher, CipherArr)
print("Original Text: ", Decipher, DecipherArr)
```
